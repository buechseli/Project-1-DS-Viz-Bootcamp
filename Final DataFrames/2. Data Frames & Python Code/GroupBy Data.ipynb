```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
movies_df = pd.read_csv("final_df.csv")
```

```python
 movies_df.groupby(["Studio", "Year"],).sum()["Total Gross"]
```

```python
movies_df = movies_df.groupby(["Year"]).mean()
del movies_df["Unnamed: 0"]
```

```python
movies_df["Total Gross"] = movies_df["Total Gross"].map("${:,.2f}".format)
movies_df["Opening"] = movies_df["Opening"].map("${:,.2f}".format)
movies_df["View Count"] = movies_df["View Count"].map("{:,}".format)
movies_df["Like Count"] = movies_df["Like Count"].map("{:,}".format)
movies_df["Dislike Count"] = movies_df["Dislike Count"].map("{:,}".format)
movies_df["Comment Count"] = movies_df["Comment Count"].map("{:,}".format)
```
