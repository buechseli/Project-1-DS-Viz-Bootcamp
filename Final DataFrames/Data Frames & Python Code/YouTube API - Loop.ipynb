```python
import json
import requests
import pandas as pd
from pprint import pprint
```

```python
df_movies = pd.read_excel('2013-2017 Box Office Data.xlsx', sheet_name=0, header=0, skiprows=None, skip_footer=0, index_col=None, names=None, usecols=None, parse_dates=False, date_parser=None, na_values=None, thousands=None, convert_float=True, converters=None, dtype=None, true_values=None, false_values=None, engine=None, squeeze=False)
df_movies
```

```python
key = "AIzaSyDJjB7jRXjdsPFeMeOJOm62lf_yk27GmY0"
```

```python
for index, row in df_movies.iterrows():
    base_url_search = "https://www.googleapis.com/youtube/v3/search?part=snippet&q=movie%2Btrailer%2B{}&key={}"
    q = row['Movie Title']
    response = requests.get(base_url_search.format(q,key)).json()
    df_movies.set_value(index,"Video ID", response['items'][0]['id']['videoId'])

df_movies
```

```python
for index, row in df_movies.iterrows():
    base_url_video = "https://www.googleapis.com/youtube/v3/videos?part=snippet,contentDetails,statistics&id={}&key={}"
    video_id = row['Video ID']
    stats = requests.get(base_url_video.format(video_id,key)).json()
    df_movies.set_value(index,"View Count", stats['items'][0]['statistics']['viewCount'])
    df_movies.set_value(index,"Video Title", stats['items'][0]['snippet']['title'])
    df_movies.set_value(index,"Like Count", stats['items'][0]['statistics']['likeCount'])
    df_movies.set_value(index,"Dislike Count", stats['items'][0]['statistics']['dislikeCount'])
    df_movies.set_value(index,"Comment Count", stats['items'][0]['statistics']['commentCount'])

df_movies
```

```python
df_movies['View Count'] = pd.to_numeric(df_movies['View Count'])
df_movies['Like Count'] = pd.to_numeric(df_movies['Like Count'])
df_movies['Dislike Count'] = pd.to_numeric(df_movies['Dislike Count'])
df_movies['Comment Count'] = pd.to_numeric(df_movies['Comment Count'])
```

```python
df_movies =df_movies[['Movie Title','View Count','Video Title','Like Count','Dislike Count','Comment Count']]
df_movies.head()
```
