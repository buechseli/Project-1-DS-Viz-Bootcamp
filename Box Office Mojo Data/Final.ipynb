```python
import pandas as pd
```

```python
movies_df = pd.read_csv('movies_data.csv')
movies_df.head(2)
```

```python
youtube_df = pd.read_csv('youtube.csv')
youtube_df.head(2)
```

```python
season_df = pd.read_csv('season.csv')
season_df.head(2)
```

```python
final_df = pd.merge(movies_df, youtube_df, left_on = movies_df['Movie Title'], right_on = youtube_df['Movie Title'])
final_df
```

```python
final_df.drop('Unnamed: 0_y', axis=1, inplace=True)
final_df.drop('Movie Title_y', axis=1, inplace=True)
final_df.drop('Unnamed: 0_x', axis=1, inplace=True)
final_df = final_df.rename(columns ={'Movie Title_x':'Movie Title'})
final_df.head()
```

```python
final_df = pd.merge(final_df, season_df, left_on = final_df['Movie Title'], right_on = season_df['Movie Title'])
final_df.head(2)
```

```python
final_df.drop('Unnamed: 0', axis=1, inplace=True)
final_df.drop('Movie Title_y', axis=1, inplace=True)
final_df = final_df.rename(columns ={'Movie Title_x':'Movie Title'})
final_df.head(3)
```

```python
final_df.dtypes
```

```python
final_df.to_csv('final_df.csv', encoding ='utf-8')
```
